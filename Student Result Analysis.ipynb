# Gender Distribution

```python
plt.figure(figsize=(9,10))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we analysed that no. of females are more than males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(9,10))
sns.heatmap(gb)
plt.title("Relationship between Parent Education and Student Performance")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent Education and Student Performance")
plt.show()
```

```python
#from above we established the relation between parent education with students performance
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent Marital Status and Student Performance")
plt.show()
```

```python
#from above we established the relation between parent marital status with students performance
```

# Distribution of Ethnic Groups

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

print(groupA, groupB, groupC, groupD, groupE)

print(groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"])


```

```python
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]

plt.figure(figsize=(10, 10))
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Groups")
plt.show()

```

```python
plt.figure(figsize = (8, 9))
ax = sns.countplot(data = df, x = "EthnicGroup")
ax.bar_label(ax.containers[0])
plt.title("Ethnic Group Count")

plt.show()
```
